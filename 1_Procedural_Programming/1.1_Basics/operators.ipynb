# **Operators**

1. ➗ Arithmetic : `+  -  *  /  //  %  **`
2. 🔧 Bitwise: `&  |  ^  ~  <<  >>`
3. 🔍 Comparison: `==  !=  >  <  >=  <=`
4. ⚡ Logical: `and  or  not`
5. 📦 Membership:`in  not in`

📊 Quantifiers : `all, any`


### ➗ Arithmetic Operators (basic math operations)

```python
a, b = 10, 3
print(a + b)    # 13 → addition
print(a - b)    # 7  → subtraction
print(a * b)    # 30 → multiplication
print(a / b)    # 3.333... → division (float result)
print(a // b)   # 3  → floor division (integer result)
print(a % b)    # 1  → remainder
print(a ** b)   # 1000 → exponentiation (10^3)
```

### 🔧 Bitwise Operators (work on binary bits)


```python
x, y = 5, 3
print(x & y)   # 1 → AND (both bits must be 1)
print(x | y)   # 7 → OR (at least one bit is 1)
print(x ^ y)   # 6 → XOR (different bits become 1)
print(~x)      # -6 → NOT (flips all bits)
print(x << 1)  # 10 → left shift (multiply by 2)
print(x >> 1)  # 2  → right shift (divide by 2)
```

### 🔍 Comparison Operators (compare values)

```python
print(a == b)  # False → equal to
print(a != b)  # True  → not equal to
print(a > b)   # True  → greater than
print(a < b)   # False → less than
print(a >= b)  # True  → greater or equal
print(a <= b)  # False → less or equal
```

### ⚡ Logical Operators (combine conditions)

```python
p, q = True, False
print(p and q)  # False → True only if both are True
print(p or q)   # True  → True if any one is True
print(not p)    # False → reverses boolean value
```

### 📦 Membership Operators (check presence in sequence)

```python
nums = [1, 2, 3]
print(2 in nums)      # True → element exists in list
print(5 not in nums)  # True → element does not exist
```

### 📊 Quantifiers (check conditions on collection)


```python
nums2 = [1, 1, 1]
print(all(n == 1 for n in nums2))  # True → all elements satisfy condition
print(any(n == 1 for n in nums2))  # True → at least one satisfies condition
```
