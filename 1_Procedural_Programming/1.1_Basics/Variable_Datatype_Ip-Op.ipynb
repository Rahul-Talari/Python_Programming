## 1️⃣ **Variables**
- Name used to store data that can change during program execution
- No datatype declaration required; datatype is decided automatically at runtime.


### NAMING RULES IN PYTHON

```python
# 1. CASE-SENSITIVE & CANNOT USE PYTHON KEYWORDS
# ------------------------------------------------
name = "John"
Name = "Jane"
print(f"name: {name}, Name: {Name}")  # Two different variables

# Reserved Keywords
import keyword
print(f"\nTotal Python Keywords: {len(keyword.kwlist)}")
print(f"Example keywords: {keyword.kwlist}")

# class=10  # SyntaxError: invalid syntax


# 2. ALLOWED CHARACTERS: LETTERS, DIGITS, UNDERSCORES; Can't start with a digit.
# ------------------------------------------------
student_name = "Alice"  # ✓ Valid
age1 = 25               # ✓ Valid

print(f"\nValid Variables: student_name={student_name}, age1={age1}")

# Uncomment to see error:
# 1student = "Error"  # ✗ Invalid: starts with number


# 3. MULTI-WORD NAMING STYLES
# ------------------------------------------------
# camelCase : first word lowercase, subsequent words capitalized
studentAge = 20
print(f"\ncamelCase: studentAge={studentAge}")

# PascalCase: each word capitalized
StudentAge = 22
print(f"PascalCase: StudentAge={StudentAge}")

# snake_case (Preferred in Python): all lowercase, words separated by underscores
student_age = 25
print(f"snake_case: student_age={student_age}")


# 4. CONSTANTS (UPPERCASE)
# ------------------------------------------------
PI = 3.14159
print(f"\nConstant: PI={PI}")
```

### Global Variable

```python
# =================
# Global Variable : Use `global` keyword to access and modify a variable inside a function
# =================

x = 10                          # Global variable

def update_global():
    global x                    # Declare x as global
    x = 20                      # Update global variable

update_global()
print(x)                        # Output: 20
```

## 2️⃣ **Data Types**

Data types are classifications of data that:
- Tell the compiler/interpreter how the data will be used
- Define what operations can be performed on the data
- Decide how much memory is required to store the data

In Python, there are several built-in data types:

### 🧠 Primitive Data Types
- Store single/simple values like numbers, text, or True/False
- Ex: `int`, `float`, `bool`, `str`, `None`

### 📦 Collection Data Types
- Store multiple values together in a single variable
- Ex: `list`, `tuple`, `dict`, `set`

## 3️⃣ **Inputs & OutPuts**
1. Automatic Datatype Widening :
    - Python automatically converts smaller datatypes into wider datatypes to avoid data loss.
    - Ex: `int + float → float`

2. Line Continuation :
    - Use backslash (\) or brackets (), [], {} to write long statements in multiple lines.

```python
# ============================================
# NOTES: IMPORTANT CONCEPTS
# ============================================

# 1. AUTOMATIC DATATYPE WIDENING : Python automatically converts smaller datatypes to wider ones to avoid data loss
# ------------------------------------------------

a = 5       # int
b = 3.2     # float
result = a + b  # int + float → float

print("Automatic Datatype Widening:")
print(f"a = {a}, type: {type(a)}")
print(f"b = {b}, type: {type(b)}")
print(f"result = {result}, type: {type(result)}")  # float


# 2. LINE CONTINUATION
# ------------------------------------------------
# Method 1: Using backslash (\)
total1 = 1 + 2 + 3 + \
         4 + 5 + 6

# Method 2: Using brackets (), [], {}
total2 = (1 + 2 + 3 +
          4 + 5 + 6)

print(f"\nLine Continuation:")
print(f"total1 = {total1}")
print(f"total2 = {total2}")


# ============================================
# STATIC INPUT - VARIABLE OPERATIONS
# ============================================

# 1. MULTIPLE ASSIGNMENT
# ------------------------------------------------

# Unpacking from list/tuple
x, y, z = [1, 2, 3]
print(f"\nUnpacking from list: x={x}, y={y}, z={z}")

# Multiple values to multiple variables
a, b, c = 1, "Am", 5.89
print(f"Multiple assignment: a={a}, b={b}, c={c}")

# Single value to multiple variables
a = b = c = 1
print(f"Single value to multiple: a={a}, b={b}, c={c}")


# 2. SWAPPING VALUES
# ------------------------------------------------
a = 10
b = 20
print(f"\nBefore swap: a={a}, b={b}")

a, b = b, a  # Swap without temporary variable
print(f"After swap: a={a}, b={b}")


# 3. DELETING VARIABLES
# ------------------------------------------------
x = 100
print(f"\nVariable x exists: {x}")

del x  # x no longer exists
# print(x)  # Uncomment to see NameError: name 'x' is not defined
print("Variable x has been deleted")
```
