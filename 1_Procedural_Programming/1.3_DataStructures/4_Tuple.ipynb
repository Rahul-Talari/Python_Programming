# **Data Structures**

| Data Structure | Characteristics |
|----------------|-----------------|
| 🔸String | Ordered + **Immutable** sequence of characters + Allows duplicates |
| 🔸List | Ordered + Mutable collection of elements + Allows duplicates |
| 🔸Tuple | Ordered + **Immutable** collection of elements + Allows duplicates |
| 🔸Set | **Unordered** + Mutable collection of unique elements + **No duplicates** |
| 🔸Dictionary | Inserted Order + Mutable collection of key-value pairs where keys are unique & values allow duplicates |

## **Tuple Operations:** 


### 🔸Creation
- Initialization

```python
tup=(1, 2, 3, 4, 5)
print(tup)
```

### 🔸Modification

- Insert : Tuples are immutable, so new elements cannot be added after creation.
- Delete : Individual elements cannot be removed because tuple size cannot change.
- Update : Existing values cannot be modified once the tuple is created.

**Note:**  To perform operations : convert tuple → list, perform changes & then convert back to tuple

### 🔸Ordering
- Sorting → Arrange elements
- Reverse → Reverse order

```python
# Sorting & Reversing Tuples

t = (40, 10, 30, 20, 50)

sorted_t = sorted(t)   # Returns new sorted list ; sort method wont work since tuples are immutable & inplace not possible
rev_t = t[::-1]        # Returns reversed copy; ; reverse method wont work since tuples are immutable & inplace not possible

print("Tuple    :", t)
print("Sorted   :", sorted_t)
print("Reversed :", rev_t)
```

### 🔸Access
- Accessing → SLICING, LOOPING, unpacking & Tuple COMPREHENSION IN PYTHON
- Search → Find element / existence check

```python
tup = (10, 20, 30, 40, 50)
n=len(tup)

# Indexing + Unpacking
print("Index :", tup[2])
a, b, c, d, e = tup
print("Unpacked :", a, b, c, d, e)

# Slicing
print("Slice +ve :", tup[0:4:1])
print("Slice -ve :", tup[-1:-n:-1])

# Looping
print("Index based Looping :")
for x in tup: print(x,end=' ')

print("\nPosition based Looping :")
for i in range(len(tup)): print(i, tup[i])

print("Index + Value Looping :")
for i, v in enumerate(tup): print(i, v)

# "Tuple comprehension" (actually generator → tuple)
print("Comp :", tuple(x**2 for x in tup))
```

### Search

```python
tup = (1, 2, 3, 2, 4, 2)

# in operator → Check existence
print(2 in tup)      # True
print(10 in tup)     # False

# Get all indices of value 2
indices = [i for i, x in enumerate(tup) if x == 2]
print(indices)
```
