# **Data Structures**

| Data Structure | Characteristics |
|----------------|-----------------|
| 🔸String | Ordered + **Immutable** sequence of characters + Allows duplicates |
| 🔸List | Ordered + Mutable collection of elements + Allows duplicates |
| 🔸Tuple | Ordered + **Immutable** collection of elements + Allows duplicates |
| 🔸Set | **Unordered** + Mutable collection of unique elements + **No duplicates** |
| 🔸Dictionary | Inserted Order + Mutable collection of key-value pairs where keys are unique & values allow duplicates |

## **Set Operations:** 


### 🔸Creation
- Initialization

```python
S=set()             # Empty Set initialization
s=set({10,20,30})   # Set initialization with values
print(f"S: {S}\ns: {s}")
```

### 🔸Modification
- Insert → Add elements
- Delete → Remove elements
- Update → Modify elements

```python
# ---------------------------------------------------------
# Insert Operations
# ---------------------------------------------------------
# Insert
s.add(40)                     # Add single element
s.update([50, 60])            # Add multiple elements from iterable (list, set, tuple)
print("After Insert :", s)

# ---------------------------------------------------------
# Delete Operations
# ---------------------------------------------------------
print("\n--- Delete ---")

new_set = s - {30}                                  # s - {x} → Remove element x and return new set  
print("New Set after removing 30:", new_set)

removed = s.pop()                                   # s.pop() → Removes & returns random element
print("Popped Element:", removed)
print("After pop():", s)

s.clear()                                           # s.clear() → Remove all elements from the set 
print("After clear():", s)

# ---------------------------------------------------------
# Update Values in Set
# ---------------------------------------------------------
print("\n--- Update Value ---")                     # Set elements cannot be updated directly : Remove old value and add new value

nums = {1, 2, 3}
nums.remove(2)
nums.add(20)

print("Updated Set:", nums)
```

### 🔸Ordering
- Sorting → Arrange elements
- Reverse → Reverse order

```python
# Sorting & Reversing Sets

s = {40, 10, 30, 20, 50}

sorted_list = sorted(s)      # Returns new sorted list; sort method wont work since sets are unordered
rev_list = list(s)[::-1]     # Convert set → list & reverse

print("Set            :", s)
print("Sorted         :", sorted_list)
print("Reverse        :", rev_list)
```

### 🔸Access

```python
# ---------------------------------------------------------
# Accessing → Looping
# ---------------------------------------------------------
s = {10, 20, 30, 40}
for item in s:                                      # Iterate using value
    print(item, end=' ')
    
    
numbers = [1, 2, 3, 4, 5, 6]
squares = {x**2 for x in numbers if x % 2 == 0}     # Set Comprehension
print("Squares:", squares)
```

### 🔸Search

```python
# ---------------------------------------------------------
# Membership Operator
# in operator → Check existence
# ---------------------------------------------------------

s = {10, 20, 30, 40}

print(20 in s)      # True
print(100 in s)     # False
```

### 🔸Set Operations

```python
# Set Operations

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("A :", A)
print("B :", B)

print("\nUnion                :", A | B)   # A.union(B)
print("Intersection         :", A & B)   # A.intersection(B)
print("Difference (A - B)   :", A - B)   # A.difference(B)
print("Symmetric Difference :", A ^ B)   # A.symmetric_difference(B)

# A ^ B = (A ∪ B) - (A ∩ B)
print("Formula Check :", (A | B) - (A & B))
```

### 🔸Set Checks

```python
# Set Checks

A = {1, 2}
B = {1, 2, 3, 4}
C = {5, 6}

print("A :", A)
print("B :", B)
print("C :", C)

print("\nA.issubset(B)   :", A.issubset(B))   # A ⊆ B
print("B.issuperset(A) :", B.issuperset(A))   # B ⊇ A
print("A.isdisjoint(C) :", A.isdisjoint(C))   # No common elements
```
