# **Data Structures**

| Data Structure | Characteristics |
|----------------|-----------------|
| 🔸String | Ordered + **Immutable** sequence of characters + Allows duplicates |
| 🔸List | Ordered + Mutable collection of elements + Allows duplicates |
| 🔸Tuple | Ordered + **Immutable** collection of elements + Allows duplicates |
| 🔸Set | **Unordered** + Mutable collection of unique elements + **No duplicates** |
| 🔸Dictionary | Inserted Order + Mutable collection of key-value pairs where keys are unique & values allow duplicates |

## **String Operations:** 


### 🔸Creation
- Initialization

```python
str1 = ""                 # Empty string
str2 = "Hello World !!"   # String with content
print(f"str1: '{str1}', str2: '{str2}'")
```

### 🔸Modification

- Insert : ❌ Not possible (Strings are immutable)
- Delete : ❌ Not possible (Cannot delete characters directly)
- Update : ❌ Not possible (Cannot modify existing characters) ✔ Convert to list → modify → join back


### 🔸Ordering
- Sorting → Arrange elements
- Reverse → Reverse order

```python
s = "hello"

print("Sorted :", "".join(sorted(s)))   # sort + join
print("Reversed:", s[::-1])             # reverse

lst = ['h','e','l','l','o']
print("Joined :", "".join(lst))         # join list to string
```

### 🔸Access: Accessing → SLICING, LOOPING & LIST COMPREHENSION IN PYTHON


```python
s = "hello world"

# Slicing
print(s[0:5:1])      # left → right
print(s[-1:-6:-1])   # right → left
print(s[::])         # full string (default slicing)

# Looping
for ch in s: print(ch)                              # value
for i in range(len(s)): print(i, s[i])              # index
for i, v in enumerate(s): print(i, v)               # index + value

# Comprehension
newlist = [ch for ch in s if ch != ' ']
print(newlist)
```

### 🔸Search → Find element / existence check

```python
s = "banana apple"

# in operator → check existence
print("a" in s)   # True

# get all index positions of 'a'
indices = [i for i, ch in enumerate(s) if ch == 'a']
print(indices)
```

### String Functions

```python
s = "  Hello World  "

# ASCII Functions
print(f"ord('a'): {ord('a')} → ASCII/Unicode value of 'a'")
print(f"chr(97): {chr(97)} → Character for ASCII/Unicode 97")

# Extraction
print(f"strip(): {s.strip()} → removes spaces from both sides")
print(f"lstrip(): {s.lstrip()} → removes spaces from left side")
print(f"rstrip(): {s.rstrip()} → removes spaces from right side")
print(f"split(): {s.split()} → splits string into list")

# Manipulation
print(f"replace(): {'hello'.replace('h','H')} → replaces substring")
print(f"lower(): {'HELLO'.lower()} → converts to lowercase")
print(f"upper(): {'hello'.upper()} → converts to uppercase")

# String Checks
print(f"isalpha(): {'abc'.isalpha()} → checks only letters")
print(f"isdigit(): {'123'.isdigit()} → checks only digits")
print(f"isnumeric(): {'123'.isnumeric()} → checks numeric values")
print(f"isalnum(): {'abc123'.isalnum()} → letters + digits only")
print(f"islower(): {'hello'.islower()} → checks lowercase")
print(f"isupper(): {'HELLO'.isupper()} → checks uppercase")

# Pattern Matching
print(f"startswith(): {'hello'.startswith('he')} → checks prefix")
print(f"endswith(): {'hello'.endswith('lo')} → checks suffix")
```
