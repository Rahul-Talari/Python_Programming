# **Data Structures**

| Data Structure | Characteristics |
|----------------|-----------------|
| 🔸String | Ordered + **Immutable** sequence of characters + Allows duplicates |
| 🔸List | Ordered + Mutable collection of elements + Allows duplicates |
| 🔸Tuple | Ordered + **Immutable** collection of elements + Allows duplicates |
| 🔸Set | **Unordered** + Mutable collection of unique elements + **No duplicates** |
| 🔸Dictionary | Inserted Order + Mutable collection of key-value pairs where keys are unique & values allow duplicates |

## **List Operations:** 


### 🔸Creation
- Initialization

```python
#=======================================================================
# List Initialization
#=======================================================================
n=10
list1 = []          # Empty List
list1 = [0] * n     # List with n elements initialized to 0
```

### 🔸Modification
- Insert → Add elements
- Delete → Remove elements
- Update → Modify elements

```python
#=======================================================================
# Insert Operations
#=======================================================================
l = [1, 2, 3]
print("Initial List:", l)

l.insert(1, 10)                   # Insert 10 at index 1
print("After insert():", l)

l.append(4)                       # Add 4 at the end  
print("After append():", l)

l.extend([5, 6])                  # Add multiple elements from iterable
print("After extend():", l)

result = l + [3, 4]               # Concatenate with another list 
print("After concatenation:", result)
```

```python

#=======================================================================
# Delete Operations
#=======================================================================
lst = [1, 2, 2, 3, 2, 4]
print("\nInitial List:", lst)

lst = [x for x in lst if x != 2]        # Remove all occurrences of 2 using list comprehension
print("After removing all 2s:", lst)    

removed = lst.pop(1)                    # Remove & return element at index 1
print("Removed Element:", removed)
print("After pop():", lst)

lst.clear()                              # Remove all elements from the list  
print("After clear():", lst)
```

```python

#=======================================================================
# Update Operation
#=======================================================================
list1 = [1, 2, 3]
print("\nInitial List:", list1)

list1[1] = 100                           # Replace element at index 1 with 100
print("After update:", list1)
```

### 🔸Ordering
- Sorting → Arrange elements
- Reverse → Reverse order

```python
#=======================================================================
# Sorting Operations
#=======================================================================

l = [4, 1, 3, 2]

print("Original List:", l)
l.sort(reverse=False)                         # Sorts original list in ascending order; default is False
print("After sort() ascending:", l)


lst = [5, 2, 8, 1]

print("Original List:", lst)
sorted_lst = sorted(lst, reverse=False)      # Returns new sorted list without modifying original; default is False
print("New Sorted List:", sorted_lst)
```

```python
#=======================================================================
# Reverse Operations
#=======================================================================
l = [1, 2, 3, 4]

print("Before reverse():", l)
l.reverse()                                  # Reverses original list in-place
print("After reverse():", l)

print("Original List:", l)
reversed_copy = l[::-1]                      # Creates reversed copy using slicing
print("Original List after slicing:", l)
print("Reversed Copy with slicing:", reversed_copy)
```

### 🔸Access
- Accessing → SLICING, LOOPING & LIST COMPREHENSION IN PYTHON
- Search → Find element / existence check

```python
# =========================================================
# LIST ACCESSING: SLICING, LOOPING & LIST COMPREHENSION IN PYTHON
# =========================================================

lst = [10, 20, 30, 40, 50, 60]

print("Original List:", lst)

# ---------------------------------------------------------
# Accessing → Indices
# list[start:stop:step]
# stop is exclusive
# Default: start = 0, stop = end, step = 1
# ---------------------------------------------------------

print("\n--- Positive Slicing ---")

print("lst[0:4:1]  :", lst[0:4:1])     # Traverse left → right
print("lst[1:5]    :", lst[1:5])       # Default step = 1
print("lst[:4]     :", lst[:4])        # Default start = 0
print("lst[2:]     :", lst[2:])        # Default stop = end


print("\n--- Negative Slicing ---")

print("lst[-1:-6:-1] :", lst[-1:-6:-1])   # Traverse right → left
print("lst[::-1]     :", lst[::-1])       # Reverse list
```

```python
# ---------------------------------------------------------
# Looping
# ---------------------------------------------------------

print("\n--- Loop Using Value ---")

for item in lst:
    print(item)


print("\n--- Loop Using Index ---")

for i in range(len(lst)):
    print(f"Index = {i}, Value = {lst[i]}")


print("\n--- Loop Using enumerate() ---")

for i, val in enumerate(lst):
    print(f"Index = {i}, Value = {val}")


# ---------------------------------------------------------
# List Comprehension
# newlist = [expression for item in list if condition]
# ---------------------------------------------------------

numbers = [1, 2, 3, 4, 5, 6]

squares = [x**2 for x in numbers if x % 2 == 0]

print("\n--- List Comprehension ---")

print("Numbers :", numbers)
print("Squares of even numbers :", squares)
```

### 🔸Search

```python
# ---------------------------------------------------------
# Membership Operator
# in operator → Check existence
# ---------------------------------------------------------

lst = [1, 2, 3, 2, 4, 2]

print(2 in lst)      # True
print(10 in lst)     # False
```
