# **Data Structures**

| Data Structure | Characteristics |
|----------------|-----------------|
| 🔸String | Ordered + **Immutable** sequence of characters + Allows duplicates |
| 🔸List | Ordered + Mutable collection of elements + Allows duplicates |
| 🔸Tuple | Ordered + **Immutable** collection of elements + Allows duplicates |
| 🔸Set | **Unordered** + Mutable collection of unique elements + **No duplicates** |
| 🔸Dictionary | Inserted Order + Mutable collection of key-value pairs where keys are unique & values allow duplicates |

## **Dictionary Operations:** 


### 🔸Creation
- Initialization

```python
from collections import defaultdict

# Normal dict
d = {}

# defaultdict examples
d_int = defaultdict(int)     # default = 0
d_float = defaultdict(float) # default = 0.0
d_str = defaultdict(str)    # default = ""
d_list = defaultdict(list)  # default = []
d_set = defaultdict(set)    # default = set()
```

### 🔸Modification
- Insert → Add elements
- Delete → Remove elements
- Update → Modify elements

```python
# Insert / Update
d["a"] = 1
d.update({"b": 2, "c": 3})

# Delete operations
val = d.pop("a")                              # remove key k and get value
d = {k: v for k, v in d.items() if k != "a"}  # remove key k
last = d.popitem()                            # remove last inserted item and return the key:value
d.clear()                                     # remove all items

# Output
print(f"Value of key 'a': {val}")             # value of 'a'
print(f"Last removed item: {last}")           # last removed item
print(f"Empty dict: {d}")                     # empty dict
```

### 🔸Access
- Accessing → Looping & Dictionary comprehension in Python


```python
# Dictionary Accessing

d = {"a": 1, "b": 2, "c": 3}

# 1. Keys
print("Keys:")
for key in d.keys():
    print(key)

# 2. Values
print("\nValues:")
for value in d.values():
    print(value)

# 3. Items (key-value pairs)
print("\nKey-Value Pairs:")
for key, value in d.items():
    print(f"Key {key} Value: {value}")
```

### 🔸Searching
- Search → Find element / existence check

```python
# Key search
print("a" in d)          # True → checks keys only

# Value search
print(2 in d.values())   # True → checks values only
```
