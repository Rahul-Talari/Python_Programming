## 🧠 Conditional Statements in Python
1. Standard format: if, elif, else
2. Short Hand if-else (Ternary Operator)


### Standarad format: if, elif, else

```python
num = 10

if num > 0:
    print("Positive")     # runs if condition is True
elif num == 0:
    print("Zero")         # runs if first condition fails and this is True
else:
    print("Negative")     # runs if all above conditions fail
```

### Short Hand if-else (Ternary Operator)

```python
result = "Positive" if num > 0 else "Non-Positive"
print(result)
```
