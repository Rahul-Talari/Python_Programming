## 🔁 **Iterative Statements in Python:** 
(for → known iterations | while → condition-based iterations)

- 1️⃣ Index-Based Iteration
- 2️⃣ Element-Based Iteration
- 3️⃣ Index + Element Iteration using enumerate()
- 🔄 while Loop
- 🔄 Loop Control Statements
    - break → stops loop completely  
    - continue → skips current iteration  
    - pass → placeholder (does nothing)

### For Loop

```python
numbers = [10, 20, 30, 40]

# 1️⃣ Index-Based Iteration
for i in range(len(numbers)):
    print(i, numbers[i])   # i = index, numbers[i] = value


# 2️⃣ Element-Based Iteration
for value in numbers:
    print(value)           # directly prints elements


# 3️⃣ Index + Element Iteration using enumerate()
for i, value in enumerate(numbers):
    print(i, value)        # i = index, value = element
```

### While Loop

```python
# 4️⃣ while Loop Iteration
i = 0
while i < len(numbers):
    print(i, numbers[i])   # index and value
    i += 1                 # increment to avoid infinite loop
```

### 🔄 Loop Control Statements (break, continue, pass)

```python
for i in range(5):

    if i == 1:
        print("break executed")
        break              # stops loop completely

    if i == 2:
        print("continue executed")
        continue          # skips current iteration

    if i == 3:
        print("pass executed")
        pass              # does nothing (placeholder)

    print(i)
```
