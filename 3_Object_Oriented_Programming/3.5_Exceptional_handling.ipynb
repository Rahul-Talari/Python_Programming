# **Exceptions in Python**

🔥 Definition: Exceptions are runtime events that disrupt the normal flow of a program.

📌 Types of Exceptions
1. **Built-in Exceptions** : SyntaxError (compile-time), ValueError, TypeError, ZeroDivisionError
2. **User-defined Exceptions**: Custom exceptions created using classes and raised using `raise`

🔄 Exception Handling Flow

- `try`     → code that may raise exception  
- `except`  → handles the exception  
- `else`    → runs if no exception occurs  
- `finally` → always runs (error or no error)

#### **Exceptional Handling**

```python
try:
    # risky code
    a = int(input("Enter number 1: "))
    b = int(input("Enter number 2: "))
    c = a / b

except ZeroDivisionError:
    print("Denominator cannot be zero")

except Exception as e:
    print("Unexpected error:", e)

else:
    print("Result:", c)

finally:
    print("Execution completed")
```

#### 🚨 Raising Exceptions

🔹 Built-in Exceptions:
Used to manually trigger standard Python errors during execution when invalid conditions occur.

- Examples:
  - ValueError → when input value is invalid
  - ZeroDivisionError → when division by zero occurs

```python
# Manually raising built-in exceptions

try:
    def divide(a, b):
        if b == 0:
            raise ZeroDivisionError("Denominator cannot be zero")
        return a / b

    print(divide(10, 0))

    def ageCheck(age):
        if age < 0:
            raise ValueError("Age cannot be negative")

    print(ageCheck(-5))

except (ZeroDivisionError, ValueError) as e:
    print("Error caught:", e)
```

#### 🧠 **Custom Exceptions**
- Custom exceptions are used to define domain-specific errors.
- We create them by inheriting from Python's built-in `Exception` class.

```python
class MyError(Exception):
    pass

x = -10
try:
    if x < 0:
        raise MyError("Negative number not allowed")
except MyError as e:
    print("Error caught:", e)
```
