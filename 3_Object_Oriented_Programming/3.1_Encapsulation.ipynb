# Encapsulation in Python

Encapsulation is the process of wrapping data (attributes) and methods into a single unit (class) and restricting direct access to data.

It is achieved using:
- Public Members
- Protected Members (_variable)
- Private Members (__variable)

```python
# Encapsulation Example

class Employee:

    def __init__(self):

        self.public = "Public"          # Public Variable: Can be accessed from anywhere inside or outside the class
        self._protected = "Protected"   # Protected Variable: Intended for internal/subclass use (accessible but convention-based)
        self.__private = "Private"      # Private Variable: Cannot be accessed directly outside the class

    def get_private(self):              
        # Public Method to access private variable
        return self.__private


obj = Employee()

print(obj.public)
print(obj._protected)
print(obj.get_private())
```
