# Inheritance in Python

Inheritance is an OOP concept that allows a class to inherit attributes and methods from another class.

🔗 Supported Types of Inheritance

```text
Single Inheritance        Multi-Level Inheritance                 Hierarchical Inheritance
---------------------     --------------------------              ---------------------------
Parent ───▶ Child         GrandParent ───▶ Parent ───▶ Child      Parent ───▶ Child1
                                                                      └──▶ Child2
```

 ❌ Not Supported Directly

```text
Multiple Inheritance      Hybrid Inheritance
---------------------     ------------------
Parent1 ──┐
           ├──▶ Child                     GrandParent
Parent2 ──┘                                    │
                                               ▼
                                            Parent
                                           /      \
                                          ▼        ▼
                                       Child1    Child2
```

📌 Concepts in Inheritance

- ✔ Parent Class
- ✔ Child Class
- ✔ Constructor Inheritance (via super())
- ✔ Method Inheritance

#### **Accessing Instance Variables & Instance Methods**


````python
class Employee:

    def __init__(self, name):                      # Parent Constructor
        self.name = name
        print("Employee Constructor Called")

    def employee_details(self):                    # Parent Method
        print(f"{self.name} works in the company")


class Developer(Employee):                        # Child Class inheriting Parent Class

    def __init__(self, name, skill):              # Child Constructor
        super().__init__(name)                    # Calling Parent Constructor using super()
        self.skill = skill
        print("Developer Constructor Called")

    def coding(self):                             # Child Method
        print(f"{self.name} writes {self.skill} code")


dev = Developer("Rahul", "Python")                # Creating Child Object

print()

dev.employee_details()                            # Accessing Parent Method
dev.coding()                                      # Accessing Child Method
````

#### **Accessing Class Variables & Class Methods**


````python
class Employee:

    company = "TCS"                 # Class Variable

    @classmethod
    def get_company(cls):           # Class Method
        return cls.company


# Accessing using Class Name
print(Employee.company)
print(Employee.get_company())
````
