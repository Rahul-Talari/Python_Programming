## **Polymorphism in Python**

Polymorphism means "many forms". It allows the same method/operator/function name to perform different behaviors depending on the object or data type.

🔹 Method Overriding (Run-time Polymorphism)

- Child class provides its own implementation of a parent class method
- Same method name and parameters

🔹 Method Overloading

- Using the same method name for different number/types of inputs
- Python does not support true method overloading directly
- Latest method definition overrides the previous one

✅ Alternative: Use default arguments (*args / default parameters)

### **Polymorphism Example**


```python
class Dog:
    def sound(self): print("Dog barks")     # Dog behavior

class Cat:
    def sound(self): print("Cat meows")     # Cat behavior

for obj in [Dog(), Cat()]:
    obj.sound()


print()
```

### **Method Overriding Example**


```python
class Animal:
    def sound(self): print("Animal sound")

class Dog(Animal):
    def sound(self): print("Dog barks")     # Method Overriding

Dog().sound()


print()
```

### **Method Overloading Alternative**

```python
class Math:

    def add(self, a, b, c=0):
        print(a + b + c)

obj = Math()

obj.add(10, 20)
obj.add(10, 20, 30)
```
