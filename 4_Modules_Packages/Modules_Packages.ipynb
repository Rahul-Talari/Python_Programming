# **Modules & Packages**

## **📄Module in Python**

A module is a single Python file (.py) containing functions, classes, and variables.

🔹 **Syntax**

- `import module_name` → imports **entire module**  
- `import module_name as alias` → imports module with shortcut name  
- `from module_name import function_name` → imports **specific function** from module  

 🔹**Built-in modules:** math, random, os, shutil, sys, datetime


```python
import math
from math import sqrt

print(math.sqrt(25))  # using module name → 5.0
print(sqrt(25))       # using direct import → 5.0
```

### 📦 **Package in Python**

- A package is a folder containing multiple modules. It is used to organize large Python projects.
- **__ init__.py** is a special file inside a package that tells Python the folder is a package and runs automatically when the package is imported.


🔹 Syntax

- `import package.module` → imports **module from package** using full path + into current namespace  
- `from package import module` → imports **module** from package  + into current namespace  
- `from package.module import function` → imports **specific function** from module  + into current namespace   


```python
# Importing from Package
from mypackage.math_utils import add

print(add(2, 3))  # calls add function → returns 5
```

#### **📦 Subpackage in Python**

A subpackage is a package inside another package, used to organize modules hierarchically into subdirectories.

🔹 Syntax

- `from package.subpackage.module import function` → imports **function** from module in subpackage + into current namespace
- `from package.subpackage import module` → imports **module** from subpackage + into current namespace
- `import package.subpackage.module` → imports **module** using full path + into current namespace

```python
# Importing from a subpackage
from mypackage.advanced.advanced_math import multiply, power

print(multiply(3, 4))  # calls multiply function → returns 12
print(power(2, 5))     # calls power function → returns 32
```
