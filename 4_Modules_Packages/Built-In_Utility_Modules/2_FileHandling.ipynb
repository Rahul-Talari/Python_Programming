## **File Handling in Python**

File handling is a fundamental operation in programming that allows you to interact with files on the filesystem.

### **Important:** Always use context managers (`with` statement) to ensure files are properly closed.

### **Text Files:**

- **Read ('r')**: Open a file for reading. The file pointer is at the beginning.- **Append Binary ('ab')**: Append binary data to existing files.

- **Write ('w')**: Open a file for writing. Creates a new file or truncates if it exists.- **Write Binary ('wb')**: Write binary data to files.

- **Append ('a')**: Open a file for appending. Writes data at the end of the file.- **Read Binary ('rb')**: Read image files, PDFs, audio, video, and other binary formats.

### **Binary Files:**

### **1. Text File Operations**

Text files are human-readable files containing characters. We'll demonstrate reading file content, counting lines/words/characters, and modifying files using write and append modes.

```python
# Read the file and print its content
file_path = 'Files/file_handling.txt'

with open(file_path, 'r') as f:
    content = f.read()
    print(content)

# Print Count lines, words, and characters
with open(file_path, 'r') as f:
    line_count, word_count, character_count = 0, 0, 0

    for line in f:
        line_count += 1                                     # Line Count
        word_count += len(line.strip().split())             # Word Count
        character_count += len(line.strip())                # Character Count (excluding leading/trailing whitespace)
        print(line.strip())

    print(f"Total number of lines: {line_count}")
    print(f"Total number of words: {word_count}")
    print(f"Total number of characters: {character_count}")
```

```python
# Overwrite the file with new content
with open('file_handling.txt', 'w') as f:
    f.write("I am writing to the file.\n")

# Append new content to the file
with open('file_handling.txt', 'a') as f:
    f.write("I am appending to the file.\n")
    
# Move the file pointer back to the beginning of the file
f.seek(0)    
```

### **2. Binary File Operations**

In this example, we'll demonstrate reading binary file data (image files).

Binary files contain non-text data such as images, audio, video, PDFs, and executable files. They cannot be read as plain text and require special handling using binary modes ('rb', 'wb', 'ab').

```python
#Read binary file of image
with open('Files/download.jpeg', 'rb') as f:
    image_data = f.read()  # Read the binary data of the image
    print(image_data[:20])  # Print the first 20 bytes of the image data
```

### **Key Takeaways:**

1. **Always use `with` statement**: Ensures files are properly closed even if errors occur.
2. **Text modes** ('r', 'w', 'a'): For readable text files
3. **Binary modes** ('rb', 'wb', 'ab'): For images, audio, video, and other binary data
4. **Write mode ('w')** overwrites existing content completely
5. **Append mode ('a')** adds content without deleting existing data
6. **Error handling**: Always check if files exist before reading to avoid FileNotFoundError
