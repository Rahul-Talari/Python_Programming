## **Logging Module**

- The `logging` module is used to track events, errors, warnings, and debugging information in programs.

| Placeholder        | Meaning              | Example                          |
|--------------------|----------------------|----------------------------------|
| `%(asctime)s`      | Timestamp            | `2026-05-18 10:30:45`            |
| `%(levelname)s`    | Log level            | `INFO`, `ERROR`, `DEBUG`         |
| `%(message)s`      | Actual log message   | `Program started`                |
| `%(name)s`         | Logger name          | `root`, `my_logger`              |
| `%(filename)s`     | File name            | `app.py`                         |
| `%(funcName)s`     | Function name        | `my_function`                    |
| `%(lineno)d`       | Line number          | `25`                             |
| `%(module)s`       | Module name          | `app`                            |
| `%(pathname)s`     | Full file path       | `C:\Projects\app.py`             |
| `%(process)d`      | Process ID           | `4321`                           |
| `%(thread)d`       | Thread ID            | `123456789`                      |

```python
import logging

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s | %(levelname)s | %(message)s',  # Format includes timestamp, level, and message
    datefmt='%Y-%m-%d %H:%M:%S',                          # Specify date/time format as YYYY-MM-DD HH:MM:SS
    handlers=[
        logging.FileHandler('Files/app.log', 'a'),   # Log to file
        logging.StreamHandler()           # Also log to console
    ]
)


# Create logger object
logger = logging.getLogger(__name__)

logging.info("Program started")

# Log messages of various severity levels
logging.debug("This is a debug message")        # Detailed debugging information: useful for diagnosing problems
logging.info("This is an info message")         # General information message   : useful for confirming that things are working as expected
logging.warning("This is a warning message")    # Warning message               : indicates a potential issue that doesn't stop the program but should be noted
logging.error("This is an error message")       # Error message                 : indicates an error that has occurred
logging.critical("This is a critical message")  # Critical/fatal error message  : indicates a serious error that may cause the program to terminate
```
