### **Built-in Utility Modules**

1. **random** : Random numbers, random selection, random shuffling, seeding

2. **os** : Directory/File/Path operations, Environment variables  
    - **shutil** : Copy/Move Directories/Files  
    - **json** : Data Serialization  
        - Python object ↔ JSON string ↔ JSON file

3. **datetime** : DateTime, Date components, Time components
4. **sys** : Python interpreter interaction, command-line arguments, module search paths

#### **1. Random**

```python
import random

# Random Numbers
print(random.random())                                    # Float between 0.0 and 1.0
print(random.uniform(1, 10))                              # Float between 1 and 10
print(random.randint(1, 10))                              # Integer between 1 and 10 (inclusive)
print(random.randrange(1, 10, 2))                         # Integer from range(start, stop, step)

# Random Selection
print(random.choice(['apple', 'banana', 'cherry']))       # Single random element
print(random.choices(['apple', 'banana', 'cherry'], k=3)) # Multiple random elements (with replacement)
print(random.sample(['apple', 'banana', 'cherry'], 3))    # Unique random elements

# Shuffle
print(random.shuffle(['apple', 'banana', 'cherry']))      # Shuffles list in-place

# Seed
print(random.seed(10))                                    # Sets seed for reproducible results

```

#### **2. OS**: File & Directory access.

```python
import os

# Current Working Directory
os.getcwd()                 # Get current directory
os.chdir(r"path")           # Change directory to desired path

# Directory Operations
os.mkdir("new_folder")      # Create directory
os.rmdir("new_folder")      # Remove empty directory
os.makedirs("a/b")          # Create nested directories
os.removedirs("a/b")        # Remove nested directories
os.listdir()                # List files/folders

# File Operations
os.rename("old.txt", "new.txt")   # Rename file
os.remove("file.txt")             # Delete file

# Path Operations
os.path.exists("file.txt")  # Check existence
os.path.isfile("file.txt")  # Check if file
os.path.isdir("folder")     # Check if directory
os.path.join("folder", "file")      # Join paths

# Environment Variables
os.getenv("PATH")           # Get environment variable: Username| API keys System

# System Information
os.name                     # Operating system name
```

##### **2.1 shutil**

```python
import shutil

# Copy Directories
shutil.copytree("src", "dest")       # Copy entire folder

# Copy Files
shutil.copy("a.txt", "b.txt")        # copy contents of file from a.txt to b.txt
shutil.copy2("a.txt", "b.txt")       # Copy contents of file from a.txt to b.txt with metadata

# Move/Rename
shutil.move("a.txt", "folder/")      # Move file/folder

# Disk Usage
shutil.disk_usage("/")               # Get Total, used, free space

# Archive Handling
shutil.make_archive("backup", "zip", "folder")      # Create zip archive of folder
shutil.unpack_archive("backup.zip", "extracted")    # Extract zip archive
```

##### **2.2 Data Serialization** 

```python
import json

data = {
    "name": "Rahul",
    "age": 25
}

json_string = json.dumps(data)		                                    #  Python object → JSON string (Serialization)
print(f"Type is :{type(json_string)} Json_string :{json_string}")		

python_obj = json.loads(json_string)	                                #  JSON string → Python object (Deserialization)
print(f"Type is :{type(python_obj)} Python_object :{python_obj}")			          


with open("data.json", "w") as file:                                    # Write serialized JSON string → JSON file
    file.write(json_string)
    
with open("data.json", "w") as file:	                                # Write deserialized Python object → JSON file
    json.dump(python_obj, file, indent=2)


with open("data.json", "r") as file:	                                # Read JSON data from file and convert into Python object
    content = json.load(file)

print(f"Type is :{type(content)} Content :{content}")                   # Output: {'name': 'Rahul', 'age': 25}
```

#### **3.datetime**

```python
from datetime import datetime, date, timedelta
import time

now=datetime.now()                                  # Current date and time
print("Current date and time:", now)              # Current date and time
#-------------------------------------
# Date Components
#-------------------------------------
print("\n\nCurrent date:", date.today())                # Current date
print("current year:", now.year)                    # Get year
print("current month:", now.month)                  # Get month
print("current day:", now.day)                      # Get day
print("current hour:", now.hour)                    # Get hour
print("current minute:", now.minute)                # Get minute
print("current second:", now.second)                # Get second

print("Time difference between now and a day ago:", now - timedelta(days=1))           # Time difference

#-------------------------------------
# Time Functions
#-------------------------------------
print("\n\nCurrent timestamp:", time.time())         # Current timestamp in seconds
print("Current local time:", time.localtime())    # Current local time
print("Current time as string:", time.ctime())        # Current time as readable string
time.sleep(2)              # Pause program for 2 seconds
```
