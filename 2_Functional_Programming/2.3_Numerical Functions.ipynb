# **Mathematical Functions Cheat Sheet**

### **1. Absolute & Rounding**
- abs(n) → absolute value
- round(n, d) → round to d decimals
- math.ceil(n) → round up
- math.floor(n) → round down

```python
import math

# 1. Absolute & Rounding
print(abs(-7))
print(round(3.1415, 2))
print(math.ceil(5.1))
print(math.floor(5.9))
```


### **2. Power & Logarithmic**
- math.sqrt(n) → square root
- math.pow(n, p) → power (float result)
- pow(n, p, mod) → modular power
- math.exp(n) → e^n
- math.log(x) → natural log
- math.log2(x) → log base 2
- math.log10(x) → log base 10

```python
import math

# 2. Power & Logarithmic
print(math.sqrt(25))
print(math.pow(2, 3))
print(pow(2, 3, 5))
print(math.exp(1))
print(math.log(10))
print(math.log2(8))
print(math.log10(100))
```


### **3. Aggregate Functions**
- min(iterable) → minimum value
- max(iterable) → maximum value
- len(iterable) → count elements
- sum(iterable) → sum of elements
- statistics.mean(data) → average
- statistics.variance(data) → variance
- statistics.stdev(data) → standard deviation

```python
import statistics

# 3. Aggregate Functions
data = [1, 2, 3]

print(min(data))
print(max(data))
print(len(data))
print(sum(data))
print(statistics.mean(data))
print(statistics.variance(data))
print(statistics.stdev(data))
```

### **4. DSA Important Functions**
- divmod(a, b) → quotient and remainder
- math.gcd(a, b) → greatest common divisor
- math.lcm(a, b) → least common multiple

```python
import math

# 4. DSA Functions
print(divmod(7, 3))
print(math.gcd(12, 8))
print(math.lcm(4, 6))
```
