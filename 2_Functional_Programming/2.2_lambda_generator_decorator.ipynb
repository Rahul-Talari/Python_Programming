### Lambda Function

- A lambda function is an anonymous (unnamed) function used for short, single-expression operations.
- It can take multiple arguments but contains only one expression.
- It automatically returns the result of the expression.

**Syntax:** lambda arguments: expression


```python
# Lambda function example
x = lambda a, b: a * b

print(x(5, 6))   # Output: 30
```

## Generator

- Generators produce values **on-demand** instead of storing them in memory.
- They **remember their state** and continue execution from where they left off.
- They are **memory efficient**, especially for large datasets or streams.

**When to use Generators:** Ideal for replacing yield with return statements {large datasets, file streams, or infinite sequences}

```python
# Generator function
def square_numbers(n):
    for i in range(1, n + 1):
        yield i * i   # produces values one by one

# Create generator object
squares = square_numbers(5)

# Iterate values
for square in squares:
    print(square)
```

### Decorator

- A decorator is a function that wraps another function to extend or modify its behavior.
- It allows adding functionality without changing the original function code.

How it works

1. Accepts the original function as input  
2. Defines an inner wrapper function  
3. Adds extra logic before/after calling the original function  
4. Returns the wrapper function  

```python
import time

# Decorator function
def timer_decorator(original_function):
    def wrapper():
        start = time.time()
        original_function()
        end = time.time()
        print("Execution Time:", end - start)
    return wrapper

# Applying decorator
@timer_decorator
def process():
    time.sleep(2)
    print("Processing...")

process()
```

```python
# Logging decorator
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function: {func.__name__}")
        print(f"Arguments: {args}, {kwargs}")
        
        result = func(*args, **kwargs)
        
        print(f"Function {func.__name__} finished")
        return result
    
    return wrapper
# Applying decorator

@log_decorator
def multiply(a, b):
    return a * b

@log_decorator
def greet(name):
    return f"Hello {name}"

# Function calls
print(multiply(3, 4))
print(greet("Alice"))
```
