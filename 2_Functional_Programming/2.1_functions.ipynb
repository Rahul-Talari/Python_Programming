### **1. What is Function**

`Function`
- A function is a reusable block of code that performs a specific task.
- It takes input (parameters), processes it, and may return a result.
- Functions improve code reusability, readability, and modularity.

`Syntax`

```python
def function_name(var1: int, var2: str, var3: bool, var4: list) -> list[str] | None:
    # code
    return values

````python
# Example function
def add_numbers(a: int, b: int) -> int:
    """Adds two integers and returns the result."""
    return a + b

# Test
print(f"sum: {add_numbers(5, 10)}")
````

### **2. Returning Multiple Values**

- When multiple values are separated by commas in a `return` statement, Python automatically packs them into a tuple.
- These values can be unpacked into variables directly.

````python
def get_person_details():
    name, age, city = "Alice", 25, "NYC"
    return name, age, city   # Python automatically returns a tuple

# Tuple unpacking
name, age, city = get_person_details()

print(f"Name: {name}\nAge: {age}\nCity: {city}")
````

### 3. Types of Function Arguments

#### **3.1 Required Arguments**  
- Required arguments must be passed in the correct positional order.
- Missing any argument will result in an error.

````python
def add(a, b):
    return a + b

print(add(3, 4))   # Output: 7
````

#### **3.2 Keyword Arguments**  
- Keyword arguments are passed using parameter names.
- Order does not matter when using keyword arguments.

````python
def introduce(name, age):
    print(f"My name is {name}, and I am {age} years old.")

introduce(age=25, name="Alice")
````

#### **3.3 Default Arguments**  
- Default arguments provide a default value if no argument is passed during function call.
- If a value is provided, it overrides the default value.

````python
def greet(name, message="Hello"):
    print(f"{message}, {name}!")

greet("Alice")                 # Uses default message
greet("Bob", "Good Morning")   # Overrides default message
````

#### **3.4 Variable-Length Arguments :** *args, **kwargs  
1. **args*: Used to pass multiple positional arguments, Internally stored as a tuple.

2. ***kwargs*: Used to pass multiple keyword arguments, Internally stored as a dictionary.

````python
# *args → multiple positional arguments
def add_all(*args):
    return sum(args)

print(add_all(1, 2, 3, 4))

# **kwargs → multiple keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Charlie", age=30, city="New York")
````
